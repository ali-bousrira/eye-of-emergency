```python
def nettoyer_texte(texte):
    texte = str(texte).lower()
    texte = re.sub(r'\d+', '', texte)
    texte = re.sub(r'[^\w\s]', '', texte)
    tokens = texte.split()
    tokens = [mot for mot in tokens if mot not in stop_words]
    tokens = [stemmer.stem(mot) for mot in tokens]
    return " ".join(tokens)
```

```python
df_nettoyer['keyword'] = df_sans_doublon['keyword'].apply(nettoyer_texte)
df_nettoyer['location'] = df_sans_doublon['location'].apply(nettoyer_texte)
df_nettoyer['text'] = df_sans_doublon['text'].apply(nettoyer_texte)
```

## Comparaison des mod√®les avec GridSearch

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

if 'text_clean' not in df_nettoyer.columns:
    df_nettoyer['text_clean'] = df_nettoyer['text'].apply(nettoyer_texte)

X_text = df_nettoyer['text_clean']
y = df_nettoyer['target']


X_text = df_nettoyer['text_clean']
y = df_nettoyer['target']

pipelines = {
    "LogReg_TFIDF": Pipeline([
        ('vect', TfidfVectorizer(max_features=5000)),
        ('clf', LogisticRegression(max_iter=1000))
    ]),
    "LogReg_BOW": Pipeline([
        ('vect', CountVectorizer(max_features=5000)),
        ('clf', LogisticRegression(max_iter=1000))
    ]),
    "RF_TFIDF": Pipeline([
        ('vect', TfidfVectorizer(max_features=5000)),
        ('clf', RandomForestClassifier())
    ]),
    "SVC_TFIDF": Pipeline([
        ('vect', TfidfVectorizer(max_features=5000)),
        ('clf', SVC())
    ])
}
```

```python

param_grids = {
    "LogReg_TFIDF": {
        'vect__ngram_range': [(1, 1), (1, 2)],
        'clf__C': [0.1, 1, 10]
    },
    "LogReg_BOW": {
        'vect__ngram_range': [(1, 1), (1, 2)],
        'clf__C': [0.1, 1, 10]
    },
    "RF_TFIDF": {
        'vect__ngram_range': [(1, 1)],
        'clf__n_estimators': [100, 200],
        'clf__max_depth': [10, 20]
    },
    "SVC_TFIDF": {
        'vect__ngram_range': [(1, 1), (1, 2)],
        'clf__C': [0.1, 1, 10],
        'clf__kernel': ['linear', 'rbf']
    }
}
```

```python

cv_results = []
best_models = {}

for name in pipelines:
    grid = GridSearchCV(pipelines[name], param_grids[name], scoring='f1', cv=3, n_jobs=-1)
    grid.fit(X_text, y)
    best_models[name] = grid.best_estimator_
    cv_results.append({
        'Model': name,
        'Best Params': grid.best_params_,
        'Best F1 Score': grid.best_score_
    })

results_df = pd.DataFrame(cv_results)
display(results_df.sort_values(by='Best F1 Score', ascending=False))
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df.sort_values(by='Best F1 Score', ascending=False),
            x='Best F1 Score', y='Model', palette='mako')
plt.title('Comparaison des mod√®les de classification de tweets')
plt.xlabel('F1-score moyen (cross-validation)')
plt.ylabel('Mod√®le + Pr√©traitement')
plt.show()
```

## Evaluation des mod√®les

```python
best_model_name = results_df.sort_values(by='Best F1 Score', ascending=False).iloc[0]['Model']
best_pipeline = best_models[best_model_name]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df_nettoyer['text_clean'], df_nettoyer['target'], test_size=0.2, stratify=df_nettoyer['target'], random_state=42
)
```

```python
best_pipeline.fit(X_train, y_train)
y_pred = best_pipeline.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score, f1_score

print(f"\nMod√®le s√©lectionn√© : {best_model_name}\n")
print("Rapport de classification :")
print(classification_report(y_test, y_pred))

acc = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy : {acc:.4f}")
print(f"Recall : {recall:.4f}")
print(f"F1-score : {f1:.4f}")
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Disaster', 'Disaster'], yticklabels=['Not Disaster', 'Disaster'])
plt.xlabel('Pr√©dit')
plt.ylabel('R√©el')
plt.title(f'Matrice de confusion : {best_model_name}')
plt.tight_layout()
plt.show()
```
